```python
import cv2 
import numpy as np
from google.colab.patches import cv2_imshow

image = cv2.imread('/content/drive/MyDrive/Images/female.tiff', 0)
inverted = np.invert(image)

cv2.imwrite('inverted.jpg', inverted)
final_frame = cv2.hconcat((image, inverted))
cv2_imshow(final_frame)
```
