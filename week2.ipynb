```python
black_risang = np.zeros(shape=(512,512,3),dtype=np.int16) # int16 untuk bit 16
plt.imshow(black_risang)                                  # np.zeros = untuk reset gambar yang telah diedit
```

```python
# perhatikan koordinat titik2 pt1 dan pt2
cv.rectangle(black_risang,pt1=(384,0),pt2=(510,150),color=(0,255,0),thickness=10)
plt.imshow(black_risang)
```

```python
cv.rectangle(black_risang,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=15)
plt.imshow(black_risang)
```

```python
cv.circle(black_risang,center=(120,100),radius=50,color=(255,0,0),thickness=8)
plt.imshow(black_risang)
```

```python
cv.line(black_risang,pt1=(0,0),pt2=(150,512),color=(255,255,255),thickness=5)
plt.imshow(black_risang)
```

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_risang,text='Hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,0),thickness=2,lineType=cv.LINE_AA)
plt.imshow(black_risang)
```

```python
black_risang2=np.zeros(shape=(512,512,3),dtype=np.int32) # np.zeros = untuk reset gambar yang telah diedit
plt.imshow(black_risang2)
```

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32) # int32 untuk format bit 32
vertices
```

```python
pts = vertices.reshape((-1,1,2)) # nilai 2 untuk menunjukan bahwa tiap titik dibuat 3 chanel yang mewakili R,G,dan B
pts                              # reshape = mengubah array yang ada dengan format  (-1,1,2)
```

```python
cv.polylines(black_risang2,[pts],isClosed=True,color=(255,0,0),thickness=5) # Polylines = tidak beraturan
plt.imshow(black_risang2)
```
