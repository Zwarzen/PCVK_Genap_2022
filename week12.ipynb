```python
def display(img, cmap='gray'):
  fig = plt.figure(figsize=(12,10))
  ax = fig.add_subplot(111)
  ax.imshow(img,cmap='gray')

reeses = cv.imread('/content/drive/MyDrive/Images/facedet/reeses_puffs.png',0)
display(reeses)
```

```python
cereals = cv.imread('/content/drive/MyDrive/Images/facedet/many_cereals.jpg',0)
display(cereals)
```

```python
# Initiate ORB detector
orb = cv.ORB_create()
kp1, des1 = orb.detectAndCompute(reeses, None)
kp2, des2 = orb.detectAndCompute(cereals, None)
bf = cv.BFMatcher(cv.NORM_HAMMING, crossCheck = True)
matches = bf.match(des1,des2)
single_match = matches[0]
single_match.distance
```

```python
len(matches)
```

```python
matches = sorted(matches, key = lambda x:x.distance)
reeses_match = cv.drawMatches(reeses, kp1, cereals, kp2, matches[:25], None, flags = 2)
display(reeses_match)
```

```python
sift = cv.xfeatures2d.SIFT_create()

kp1, des1 = sift.detectAndCompute(reeses, None)
kp2, des2 = sift.detectAndCompute(cereals, None)
bf = cv.BFMatcher()
matches = bf.knnMatch(des1, des2, k=2)

good = []

for match1, match2 in matches:
  if match1.distance < 0.75*match2.distance:
    good.append([match1])

len(good)
```

```python
len(matches)
```

```python
sift_matches = cv.drawMatchesKnn(reeses, kp1, cereals, kp2, good, None, flags = 2)
display(sift_matches)
```
